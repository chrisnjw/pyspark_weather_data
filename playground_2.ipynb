Features that I would pick + encoding method:

TIME: 

* DATE: Obtain hour and day_of_year from it. Perform cyclic encoding (sin and cos) on both hour and day_of_year

LOCATION: 

* LATITUDE and LONGITUDE: Perform Spherical 3D Embedding on them (change them to cartesian coordinates on the map):
```
phi = np.deg2rad(latitude); lam = np.deg2rad(longitude)
x = np.cos(phi) * np.cos(lam)
y = np.cos(phi) * np.sin(lam)
z = np.sin(phi)
```

* ELEVATION: No transformation

OTHER:

Wind Condtion: 
* WND: Only take 4th value (which is the wind speed in tenths of m/s, so 0057 is 5.7 m/s, do this transformation appropriately), make sure to only select values where the 5th value is either 1 or 5.

Cloud Cover/Ceiling
* CIG: Take 1st value which represents the cloud ceiling height, only use value where second value is 1 or 5 (also first value 01524 is 1524 m)

Visibility 
* VIS: Take 1st value which represents the horizontal visibility distance (used to represent haze and fogs), only use value where second value is 1 or 5 (also first value 016000 is 16000 m)

Dew Point
* DEW: Take 1st value which represents the dew point which represents atmospheric moisture, only use value where second value is 1 or 5 (also the +0122 actually means 12.2, look at the signs too)

Sea Level Pressure:
* SLP: Take 1st value which represents sea-level pressure in hPa with one decimal place (so 10038 is 1003.8 hPa), only use it if second value is 1 or 5.
